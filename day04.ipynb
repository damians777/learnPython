## 1. 조건문 conditional statement
조건에 따라 실행할 코드블록을 달리 구성하여 프로그램의 흐름을 제어할 수 있다.  
syntex :   
<pre>    
if 조건식 :   
    실행문장1  
    실행문장2  
else :   
    실행문장1  
    실행문장2  
</pre>    

```python
birthyear = int(input("생년은?"))
age = 2021 - birthyear
if age>=20 :
    print("성년, 입장가능")
    print("성년이래")
else :
    print("미성년, 입장불가")
    print("미성년이래")
print("프로그램 종료...")    
```

```python
# 연습문제 1. 숫자를 입력하여  홀수 인지 짝수 인지 판단하여 출력하는
# 프로그램을 작성해보세요.

a = int(input("숫자입력"))

if a % 2 == 0 :
    print("짝수")
else :
    print("홀수")
```

```python
# 연습문제 2. 숫자를 입력받아 3과 5의 공배수인지 판단하여 출력하는
# 프로그램을 작성해보세요.

a = int(input("숫자입력"))

if a % 3 == 0 and a % 5 == 0 :
    print(f"숫자 {a}는 3과 5의 공배수입니다.")
elif a %3 == 0 :  #3의 배수인 경우
    print(f"숫자 {a}는 3의 공배수입니다.")
elif a %5 == 0 :  #5의 배수인 경우
    print(f"숫자 {a}는 5의 공배수입니다.")
else :
    print(f"숫자 {a}는 3과 5의 배수도, 공배수도 아닙니다.")
print("프로그램 종료..")
```

## 2. 반복문 literative statement(loop control)


```python
for i in range(1,101) :
    print(i)
```

```python


for i in range(0, 11) :
    print(i)
```

```python
for ch in "hello python" :
    print(ch)
```

```python
words = ["korea", "south", "python", "language"]
for word in words :
    print(word)
    print("----")
    for ch in word :
        print(ch)
```

```python
cnt = int(input("몇번이나 반복할까요?"))
for i in range(cnt) :
    print("만세", i)
```
