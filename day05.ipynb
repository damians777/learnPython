```python
# unit 20
# 단계 1 : 1부터 100까지 숫자를 출력
for i in range(1, 101) :
    print(i, end=" ") # end를 통해 줄바꿈 하지 않고 출력.
```

```python
# 단계 2 :
for i in range(1, 101) :
    print(i)
    if i % 3 == 0 :
        print("Fizz", end=" ")
    else :
        print(i, end=" ")
```

```python
# 단계 3 :
for i in range(1, 101) :
    print(i)
    if i % 3 == 0 :
        print("Fizz", end = " ")
    elif i% 5 == 0 :
        print("Buzz", end=" ")
    else :
        print(i, end = " ")
```

```python
# 단계 4 :
for i in range(1, 101) :
    print(i)
#     if i % 3 == 0 and i % 5 == 0 :
#         print("FizzBuzz", end=" ")
    if i % 15 == 0 :
        print("FizzBuzz", end=" ")
    elif i % 3 == 0 :
        print("Fizz", end = " ")
    elif i% 5 == 0 :
        print("Buzz", end=" ")
    else :
        print(i, end = " ")
```
